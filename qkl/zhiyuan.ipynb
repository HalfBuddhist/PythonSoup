### 公元纪年转换

```python
# fibonacci
import math
import numpy as np
def fibonacci(n):
    f = math.sqrt(5)
    return round((math.pow(((1+f)/2), n) - math.pow(((1-f)/2), n))/f)

# Y axis: 2^n + n
def ordinate(n): 
    return round(math.pow(2, n-1) + n - 1)

# Sn of fibo
def sn_x(n):
    return 2 * fibonacci(n) + fibonacci(n-1) - 1

# Sn of y: 2*(2^n-1) + n/2*(n+1)
def sn_y(n):
    return round(2 * (math.pow(2, n) - 1) + n/2*(n+1))

%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib
# print(matplotlib.matplotlib_fname()) ## 输出字体文件夹
# (2020, 3620)
N = 15
x10 = [sn_x(e+1) for e in range(N)]
y10 = [ordinate(e+1) for e in range(N)]
for i in range(N):
    print("%s: %s, %s" % (i, x10[i], y10[i]))
plt.figure(num=3, figsize=(16, 10))
plt.rcParams['font.family'] = 'SimHei' 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.plot(x10, y10)
#画曲线1
# plt.plot(x, y1)
#画曲线2
# plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='--')
#设置坐标轴范围
# plt.xlim((-5, 5))
# plt.ylim((-2, 2))
#设置坐标轴刻度
my_x_ticks = [e*100 for e in range(17)]
my_y_ticks = [e*1000 for e in range(18)]
my_y_ticks[0] = 1
plt.xticks(my_x_ticks, [e + 2020 for e in my_x_ticks])
plt.yticks(my_y_ticks)
#设置坐标轴名称
plt.xlabel('公元纪年')
plt.ylabel('智元纪年')
plt.grid()  # 生成网格
plt.show()

# (2020, 2300)
N = 12
x10 = [sn_x(e+1) for e in range(N)]
y10 = [ordinate(e+1) for e in range(N)]
for i in range(N):
    print("%s: %s, %s" % (i, x10[i], y10[i]))
plt.figure(num=3, figsize=(16, 10))
plt.rcParams['font.family'] = 'SimHei' 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.plot(x10, y10)
#画曲线1
# plt.plot(x, y1)
#画曲线2
# plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='--')
#设置坐标轴范围
# plt.xlim((-5, 5))
# plt.ylim((-2, 2))
#设置坐标轴刻度
my_x_ticks = [e*20 for e in range(20)]
my_x_ticks[0] = 1
my_y_ticks = [e*100 for e in range(22)]
my_y_ticks[0] = 1
plt.xticks(my_x_ticks, [e + 2019 for e in my_x_ticks])
plt.yticks(my_y_ticks)
#设置坐标轴名称
plt.xlabel('公元纪年')
plt.ylabel('智元纪年')
plt.grid()  # 生成网格
plt.show()


# (2020-3000)
N = 14
x10 = [sn_x(e+1) for e in range(N)]
y10 = [ordinate(e+1) for e in range(N)]
for i in range(N):
    print("%s: %s, %s" % (i, x10[i], y10[i]))
plt.figure(num=3, figsize=(16, 10))
plt.rcParams['font.family'] = 'SimHei' 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.plot(x10, y10)
#画曲线1
# plt.plot(x, y1)
#画曲线2
# plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='--')
#设置坐标轴范围
# plt.xlim((-5, 5))
# plt.ylim((-2, 2))
#设置坐标轴刻度
my_x_ticks = [e*50 for e in range(21)]
# my_x_ticks[0] = 1
my_y_ticks = [e*400 for e in range(22)]
my_y_ticks[0] = 1
plt.xticks(my_x_ticks, [e + 2020 for e in my_x_ticks])
plt.yticks(my_y_ticks)
#设置坐标轴名称
plt.xlabel('公元纪年')
plt.ylabel('智元纪年')
plt.grid()  # 生成网格
plt.show()

N = 30
x10 = [sn_x(e+1) for e in range(N)]
y10 = [ordinate(e+1) for e in range(N)]
from bisect import bisect_left
def query(n):
    """Transfer from x to y
    """
    if n < 2020:
        print("year should larger than 2020!")
        exit(0)
    sn = n - 2019
    index = bisect_left(x10, sn)
    return math.floor(y10[index-1] + (sn-x10[index-1])/(x10[index]-x10[index-1])*(y10[index]-y10[index-1]))

#sample query
for i in range(300):
    print("%s - %s" % ((2020+i), query(2020+i)))

```
